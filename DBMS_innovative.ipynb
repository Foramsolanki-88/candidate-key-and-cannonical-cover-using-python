```python
from tkinter import *
import tkinter as tk
import itertools 
from itertools import chain, combinations


root=tk.Tk()  #initializes this interpreter and creates the root window
label1 = Label( root, bg='black')
label1.place(x = 0,y = 0)

frame1=Frame(root,padx=20,pady=20,borderwidth=5)
frame1.grid()

frame2=Frame(root,padx=20,pady=20,borderwidth=5)
frame2.grid()


table=[]
prime=[]
non_prime=[]
lhs=[]
rhs=[]
candidate_keys=[]

# label to show miner details
label1=tk.Label(frame1,text="Enter attribute:(A,B,C,D) like wise:")
label1.grid(row=0,column=0,columnspan=4,padx=10,pady=10)

# input pannel
e=tk.Entry(frame1,width=50,borderwidth=5)
e.grid(row=1,column=0,columnspan=4,padx=10,pady=10)


# main function to find candidate key 
def find_cadidate_key(table,prime,non_prime,lhs,rhs,candidate_keys):
    attribute=e.get()
    table=attribute.split(',')
    table_={table[i] for i in range(0,len(table))} #schema in set
    fds=[]
    fo=open("example.txt","r+")
    for line in fo:
        fd = line.strip().replace(" ", "").split('->',1)
        l=fd[0]
        r=fd[1]
        lhs.append(l)
        rhs.append(r)
        if len(l)>1:#if multiple attributes
            l=list(l)
            l=tuple(l)
        if len(r)>1:#if multiple attributes
            r=list(r)
            r=tuple(r)#make tuple then append
        fds.append((l,r))
        
    
    fds_={fds[i] for i in range(0,len(fds))}
    candi=compute_all_candidate_keys(table_, fds_)#pass set of attributes and set of FDs
    candidate_keys1=[list(x) for x in candi]   
    fo.close()
    for i in candidate_keys1:
        for j in i:
            prime.append(j)
            
    prime=set(prime)
    prime=list(prime)
    
    for i in table:
        if i in prime:
            continue
        else:
            non_prime.append(i)
            
    for i in candidate_keys1:
        st=''
        for k in i:
            st+=k
        candidate_keys.append(st)
    
#     print(lhs)
#     print(rhs)
#     print(prime)
#     print(non_prime)
#     print(table)
    key.insert(tk.INSERT,candidate_keys)
    
    

def powerset(iterable):
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1))

def compute_closure(X, F):
    X_plus = X
    change = True
    while change:
        change = False
        for w,z in F:
            if type(w) == tuple:
                w_set = set(w)
            else:
                w_set = {w}
            if type(z) == tuple:
                z_set = set(z)
            else:
                z_set = {z}            
            if w_set.issubset(X_plus) \
                and not z_set.issubset(X_plus):#if x attributes are subset of closure and y attributes are not add them to the closure
                X_plus = X_plus.union(z_set)
                change = True
    return X_plus

def compute_candidate_key(X, R, F):
    # X is a superkey
    T = set()
    remove = True
    while remove:
        for i in X:
            X_inter = compute_closure(X.difference({i}), F)# if same closure after removing attribute than that subset is candidate key
            if X_inter == R:
                X = X.difference({i})
            else:
                remove = False
    T = T.union(X)
    return T

def compute_all_candidate_keys(R, F):
    candidate_keys = []
    F_LHS = set()
    F_RHS = set()
    for w, z in F:
        if type(w) == tuple:# for multiple attributes
            w_set = set(w)
        else:
            w_set = {w}#single attribute
        if type(z) == tuple:
            z_set = set(z)
        else:
            z_set = {z}
        F_LHS = F_LHS.union(w_set)#x attributes
        F_RHS = F_RHS.union(z_set)#y attributes
    powerset_X = list([set(x) for x in powerset(F_LHS)])#all subsets of x attributes
    for s in powerset_X:
        # whether set s is a superkey
        if compute_closure(s,F) == R:
            X_superkey = s
            # closure = R and not in candidate key ---> a potential candidate key
            if compute_closure(X_superkey, F) == R \
                and X_superkey not in candidate_keys:
                candidate_keys.append(
                   compute_candidate_key(X_superkey, R, F))
    candidate_keys = set([frozenset(x) 
                          for x in candidate_keys 
                          if x != set()])
    return candidate_keys


# main function to find canonical cover
def find_canonical_cover():
    attribute=e.get()
    table=attribute.split(',')
    print(table)
    FD=[]
    fo=open("example.txt","r+")
    for line in fo:
        fd = line.strip().replace(" ", "").split('->',1)
        l=list(fd[0])
        r=list(fd[1])
        FD.append([l,r])
        
    fo.close()    
    print(FD)
    cover=canonical_cover(table,FD)
    for i in cover:
        cover=i
    print(cover)
    print("FDs")
    for i in FD:
        l=i[0]
        r=i[1]
        st=''
        for i in l:
            st+=str(i)
                
        st+='->'
        for i in r:
            st+=str(i)
        print(st)
        que.insert(tk.INSERT, '\n' + st)
    print("Canonical Cover")
    for i in cover:
        l=i[0]
        r=i[1]
        st=''
        for i in l:
            st+=str(i)
                
        st+='->'
        for i in r:
            st+=str(i)
        print(st)
        ans.insert(tk.INSERT, '\n' + st)
        

def closure(table, FD, S):
    if not (is_subset(S, table)):
        return [] 
    result = list(S)
    changed = True
    while (changed):
        changed = False
        for fd in FD:
            LHS = fd[0]
            RHS = fd[1]
            if (is_subset(LHS, result)):
                for att in RHS:
                    if (att not in result):
                        result.append(att)
                        changed = True
    return sorted(result)

def is_subset(X, Y):
    return set(X).issubset(set(Y)) 

def subsets(table):
    x = len(table)
    masks = [1 << i for i in range(x)]
    result = []
    for i in range(1, 1 << x):
        r = []
        for mask, ss in zip(masks, table):
            if i & mask:
                r.append(ss)
        result.append(r)  
    result.sort() 
    result.sort(key=len) 
    return result

def canonical_cover(table, FD):
    FD1 = step1(FD)
    FD2 = step2(table, FD, FD1)
    FD3 = []
    for fd in FD2:
        FD3.append(step3(table, fd))
    return FD3

def step1(FD):
    FD1 = []
    for fd in FD:
        rhs_size = len(fd[1])
        if (rhs_size <= 1):
            FD1.append(fd)
        else:
            for att in fd[1]:
                FD1.append([fd[0], [att]])
    return FD1


def step2(table, FD, FD1):
    all_possible_fds = []
    FD2 = []
    for fd in FD1:
        possible_fds = []
        lhs_size = len(fd[0])
        if (lhs_size <= 1):
                possible_fds.append(fd)
        else:
            found = False
            for att in fd[0]:
                if fd[1][0] in closure(table, FD, [att]):
                    found = True
                    possible_fds.append([[att], [fd[1][0]]])
            if not found:
                possible_fds.append(fd)
        all_possible_fds.append(possible_fds)

    for tup in itertools.product(*all_possible_fds):
        FD2.append(list(tup))
    return FD2


def step3(table, FD2):
    FD3 = remove_duplicates(FD2)
    FD3 = remove_redundant_fd(table, FD3)
    FD3 = remove_extraneous_att(table, FD3)

    return FD3

def remove_duplicates(FD):
    FD2 = list(FD)
    FD2.sort()
    FD2 = list(FD2 for FD2,_ in itertools.groupby(FD2))
    return FD2

def remove_redundant_fd(table, FD):
    for fd in FD:
        FD2 = list(FD)
        FD2.remove(fd)
        if is_subset(fd[1], closure(table, FD2, fd[0])):
            FD.remove(fd)
    return FD

def remove_extraneous_att(table, FD):
    for fd in FD:
        for att in fd[0]:
            if is_subset(att, closure(table, FD, [n for n in fd[0] if n != att])):
                fd[0].remove(att)   
    return FD


key=tk.Text(frame1,width=38,height=1,borderwidth=5)
key.grid(row=2,column=0,columnspan=4,padx=10,pady=10)

#form=tk.Text(frame1,width=38,height=1,borderwidth=5)
#form.grid(row=3,column=0,columnspan=4,padx=10,pady=10)

# text to print given functional dependencies        
que=tk.Text(frame1,width=20,height=15,borderwidth=5)
que.grid(row=4,column=0,padx=10,pady=10)

# text to print reduced functional dependencies by canonical cover
ans=tk.Text(frame1,width=20,height=15,borderwidth=5)
ans.grid(row=4,column=1,padx=10,pady=10)

# button to find candidate key
button1=tk.Button(frame2,text="Candidate Key",command=lambda:find_cadidate_key(table,prime,non_prime,lhs,rhs,candidate_keys),borderwidth=5,fg='black',bg='cyan')
button1.grid(row=0,column=0,padx=10,pady=10)




# button to find canonical cover
button3=tk.Button(frame2,text="Canonical Cover",command=find_canonical_cover,borderwidth=5,fg='black',bg='yellow')
button3.grid(row=0,column=2,padx=10,pady=10)


# main display
root.mainloop()
```
